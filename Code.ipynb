```python
distance = {
    "Sepang Kelod": {"Sepang": 6, "Tista": 6},
    "Sepang": {"Puncaksari": 5, "Tista": 4, "Sepang Kelod" : 6},
    "Tista": {"Puncaksari": 6, "Bongancina": 3, "Sepang": 4, "Sepang Kelod": 6},
    "Bongancina": {"Puncaksari": 2, "Tista": 3},
    "Puncaksari": {"Telaga": 3, "Titab": 5, "Subuk": 4, "Bongancina": 2, "Tista": 7, "Sepang": 5},
    "Telaga": {"Titab": 2, "Puncaksari": 3},
    "Titab": {"Busungbiu": 6, "Kekeran": 2, "Subuk": 4, "Puncaksari": 5, "Telaga": 2},
    "Subuk": {"Kedis": 4, "Kekeran": 6, "Tinggarsari": 5,"Titab": 4, "Puncaksari": 3},
    "Busungbiu": {"Pelapuan": 7, "Kekeran": 4, "Titab": 6},
    "Kekeran": {"Kedis": 2, "Pelapuan": 4, "Busungbiu": 4, "Titab": 2, "Subuk": 6},
    "Kedis": {"Bengkel": 3, "Pelapuan": 6, "Kekeran": 2, "Subuk": 4, "Tinggarsari": 7},
    "Tinggarsari": {"Bengkel": 6, "Kedis": 7, "Subuk": 5},
    "Pelapuan": {"Bengkel": 8, "Kedis": 6, "Kekeran": 4, "Busungbiu": 7},
    "Bengkel": {"Umajero": 6, "Tinggarsari": 6, "Kedis": 3, "Pelapuan": 8},
    "Umajero": {"Bengkel": 6}
}

road_condition = {
    "Sepang Kelod": {"Sepang": "rusak sedang", "Tista": "rusak sedang"},
    "Sepang": {"Pucaksari": "baik", "Tista": "rusak" , "Sepang Kelod": "rusak sedang"},
    "Tista": {"Puncaksari": "baik", "Bongancina": "rusak", "Sepang": "rusak", "Sepang Kelod": "rusak sedang"},
    "Bongancina": {"Puncaksari": "baik", "Tista": "rusak"},
    "Puncaksari": {"Telaga": "rusak sedang", "Titab": "rusak", "Subuk": "rusak", "Bongancina": "baik", "Tista": "baik", "Sepang": "baik"},
    "Telaga": {"Titab": "baik", "Puncaksari": "rusak sedang"},
    "Titab": {"Busungbiu": "rusak sedang", "Kekeran": "rusak sedang", "Subuk": "rusak", "Puncaksari": "rusak", "Telaga": "baik"},
    "Subuk": {"Kedis": "rusak sedang", "Kekeran": "baik", "Tinggarsari": "baik", "Titab": "rusak", "Puncaksari": "rusak"},
    "Busungbiu": {"Pelapuan": "baik", "Kekeran": "rusak", "Titab": "rusak sedang"},
    "Kekeran": {"Kedis": "baik", "Pelapuan": "baik", "Busungbiu": "rusak", "Titab": "rusak sedang", "Subuk": "baik"},
    "Kedis": {"Bengkel": "baik", "Pelapuan": "rusak", "Kekeran": "baik", "Subuk": "rusak sedang", "Tinggarsari": "rusak sedang"},
    "Tinggarsari": {"Bengkel": "baik", "Kedis": "rusak sedang", "Subuk": "baik"},
    "Pelapuan": {"Bengkel": "rusak", "Kedis": "rusak", "Kekeran": "baik", "Busungbiu": "baik"},
    "Bengkel": {"Umajero": "rusak sedang", "Tinggarsari": "baik", "Kedis": "baik", "Pelapuan": "rusak"},
    "Umajero": {"Bengkel": "rusak sedang"}
}

condition_mapping = {
    "baik": 1,
    "rusak sedang": 2,
    "rusak": 4
}
```

```python
def find_route(start, destination, route, total_distance, road_track, condition_value):
    if start == destination:
        return [(
            " -> ".join(route),
            total_distance,
            road_track.strip("-"),
            condition_value
        )]

    results = []

    if start not in distance:
        return []

    for city in distance[start]:
        if city not in route:
            d = distance[start][city]
            cond = road_condition[start][city]
            cval = condition_mapping[cond]

            route.append(city)

            results.extend(
                find_route(
                    city,
                    destination,
                    route,
                    total_distance + d,
                    road_track + "-" + cond,
                    condition_value + cval
                )
            )

            route.pop()

    return results
```

```python
start = input("Masukkan titik asal: ")
destination = input("Masukkan titik tujuan: ")

travel_results = find_route(start, destination, [start], 0, "", 0)

if not travel_results:
    print("Tidak ada rute tersedia.")
else:
    paths, distances, road_conditions, condition_values = zip(*travel_results)

    best_condition_index = condition_values.index(min(condition_values))
    shortest_distance_index = distances.index(min(distances))

    print("\n===== HASIL =====")

    if best_condition_index == shortest_distance_index or \
       distances[best_condition_index] == distances[shortest_distance_index]:
        print("\nRUTE TERBAIK:")
        print("Rute:", paths[best_condition_index])
        print("Jarak:", distances[best_condition_index], "km")
        print("Kondisi:", road_conditions[best_condition_index])

    elif condition_values[best_condition_index] == condition_values[shortest_distance_index]:
        print("\nRUTE TERBAIK:")
        print("Rute:", paths[shortest_distance_index])
        print("Jarak:", distances[shortest_distance_index], "km")
        print("Kondisi:", road_conditions[shortest_distance_index])

    else:
        print("\nRUTE KONDISI TERBAIK:")
        print("Rute:", paths[best_condition_index])
        print("Jarak:", distances[best_condition_index], "km")
        print("Kondisi:", road_conditions[best_condition_index])

        print("\nRUTE TERPENDEK:")
        print("Rute:", paths[shortest_distance_index])
        print("Jarak:", distances[shortest_distance_index], "km")
        print("Kondisi:", road_conditions[shortest_distance_index])

    print("\n===== REKOMENDASI ALTERNATIF =====")

    recommendation_count = 0
    for i in range(len(travel_results)):
        if recommendation_count < 4 and \
           distances[i] > distances[shortest_distance_index] and \
           distances[i] < distances[shortest_distance_index] + 6 and \
           condition_values[i] > condition_values[best_condition_index] and \
           condition_values[i] < condition_values[best_condition_index] + 3:

            recommendation_count += 1
            print("\nAlternatif:", recommendation_count)
            print("Rute:", paths[i])
            print("Jarak:", distances[i], "km")
            print("Kondisi:", road_conditions[i])

    if recommendation_count == 0:
        print("Tidak ada rute alternatif.")
```
